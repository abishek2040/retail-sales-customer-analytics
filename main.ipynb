```python
# Importing all the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the Excel file
data = pd.read_excel('online_retail_II.xlsx', sheet_name='Year 2010-2011')

# Displaying the first few rows of the dataset
data.head()

# Displaying the summary statistics of the dataset
data.describe()
```

```python
# Removing the missing rows in the 'Description' column
data = data.dropna(subset=['Description'])
```

```python
# Removing the missing rows in the 'Customer ID' column
data_customers = data.dropna(subset=['Customer ID'])
```

```python
# Visualising the data first to check for any outliers
# Creating a boxplot to check for outliers in the 'Quantity' column
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Quantity'])
plt.title('Boxplot of Quantity')
plt.show()
```

```python
# Converting the invoice date to datetime format
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Extracting the month and year from the invoice date
data['Month'] = data['InvoiceDate'].dt.month
data['Year'] = data['InvoiceDate'].dt.year
```

```python
# Creating a new revenue column by multiplyiing the quantity and price
data['Revenue'] = data['Quantity'] * data['Price']

print(data.head())
```

```python
data["Invoice"] = data["Invoice"].astype(str)
data["Cancelled"] = data["Invoice"].str.startswith("C")

print(data.head())
```
